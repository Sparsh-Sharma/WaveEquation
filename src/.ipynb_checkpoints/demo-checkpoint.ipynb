```julia
using WaveEquation
using Plots
```

```julia
g0(x) = exp(-40*(x-0.6)^2)
dgdt0(x) = 0.0
resolution = 500
initialstate = [discretize(g0, resolution), discretize(dgdt0, resolution)];
```

```julia
xs = range(0, 1; length=1001)
plot(xs, g0.(xs))
```

```julia
dt = 0.0001
n = 10000
result = solve(waveEQ, initialstate, dt, n);
```

```julia
xs = [i/1000 for i in 0:1000];
plotresult(xs, result, i) = plot(xs, [evaluate(result[i][1], x) for x in xs], 
    label="", 
    lw=2, 
    xlabel="x", 
    ylabel="A", 
    title="t = $((i-1)*dt)",
    ylim=[0,1]
)
```

```julia
plotresult(xs, result, 100)
```

```julia
anim = @animate for i=1:50:length(result)
    plotresult(xs, result, i)
end
```

```julia
gif(anim, fps=30)
```

```julia
potential_energy = [quad_integrate(ddx(result[i][1])^2) for i in 1:length(result)];
kinetic_energy = [quad_integrate(result[i][2]^2) for i in 1:length(result)];
```

```julia
plot(1:length(result), [potential_energy, kinetic_energy, potential_energy+kinetic_energy])
```
